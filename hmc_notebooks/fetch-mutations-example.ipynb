```python
import json
import requests

# Grab the list of Sample IDs from your study of interest
study = 'pptc_2019_all'
url = f'https://www.cbioportal.org/api/sample-lists/{study}'

response = requests.get(url)

# Will probably need to have this be part of an try except or have a status code check in your formal code, but for now just save the sampleIds field from the response to a variable
sampleIds = response.json()['sampleIds']


# Grab mutations from a study using the list of sampleIds
molecular_profile_id = 'pptc_2019_mutations'
url = f'https://www.cbioportal.org/api/molecular-profiles/{molecular_profile_id}/mutations/fetch?projection=SUMMARY&pageSize=10000000&pageNumber=0&direction=ASC' # TODO: I grabbed this URL directly from the POST api page. Everything that comes after the ? is an optional parameter I believe and as long as you include the fetch part this should work maybe

# Format our sample IDs as a json type object
data = {"sampleIds": sampleIds}

# Tell the cBioPortal resource to expect JSON
headers = {'Content-Type': 'application/json'}

response = requests.post(url,data=json.dumps(data), headers=headers)

# Check it out, we got mutations!
print(f'Mutations Found: {len(response.json())}')
response.json()
```
